```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('test_images/0 (239).jpg',0)      # 0 signifies gray image
height,width=img.shape
plt.imshow(img,cmap='gray')
```

```python
imgFlatten=img.reshape(height*width,1)
```

```python
from sklearn.cluster import KMeans

model=KMeans(n_clusters=6)
model.fit(imgFlatten)
labels=model.labels_
```

```python
labels2D=labels.reshape(height,width)
```

```python
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
import joblib

joblib.dump(model,'brain_tumor_kmeans_model.sav')
```
